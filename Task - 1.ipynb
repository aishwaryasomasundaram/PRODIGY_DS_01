```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:/Users/HP/Downloads/football results.csv", parse_dates=['date'])
df=df.head(300)
# Create a bar plot
# Group by country and plot the mean home score
df.groupby('country')['home_score'].mean().plot(kind='barh', color='green')  
plt.title('Football Results')
plt.xlabel('Country')
plt.ylabel('Average Score')  
plt.show()
```
